```python
def multiply (a,b):
    print("Multiply: ",a*b)

def add (a,b):
    print ("Add: ",a+b)

def subtract (a,b):
    print ("Subtract: ",a-b)

def divide (a,b):
    print ("Divide: ",a/b)

def calculator(a,b,operation):
    if operation == "+":
        add(a,b)
    elif operation == "-":
        subtract(a,b)
    elif operation == "*":
        multiply(a,b)
    elif operation == "/":
        divide(a,b)
    else:
        print("Invalid Operation")

calculator(5,10,"+")


```
